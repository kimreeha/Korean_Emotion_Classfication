```python
import os
import shutil

## 파일명을 감정N.jpg 형태로 일괄 변경
## train, val, test 디렉토리를 생성, 감정0~239, 240~269, 270~299 순으로 8:1:1 비율로 복사하여 학습에 사용함 
```

```python
# 경로 설정
data_dir = '/content/drive/MyDrive/facial/기쁨/'
data_dir1 = '/content/drive/MyDrive/facial/당황/'
data_dir2 = '/content/drive/MyDrive/facial/분노/'
data_dir3 = '/content/drive/MyDrive/facial/불안/'
data_dir4 = '/content/drive/MyDrive/facial/상처/'
data_dir5 = '/content/drive/MyDrive/facial/슬픔/'
data_dir6 = '/content/drive/MyDrive/facial/중립/'

# 파일명 일괄 변경하기
file_names = os.listdir(data_dir)
file_names1 = os.listdir(data_dir1)
file_names2 = os.listdir(data_dir2)
file_names3 = os.listdir(data_dir3)
file_names4 = os.listdir(data_dir4)
file_names5 = os.listdir(data_dir5)
file_names6 = os.listdir(data_dir6)

i = 0
for name in file_names:
  src = os.path.join(data_dir, name)
  dst = '기쁨' + str(i) + '.jpg'
  dst = os.path.join(data_dir, dst)
  os.rename(src,dst)
  i += 1

i = 0
for name in file_names1:
  src = os.path.join(data_dir1, name)
  dst = '당황' + str(i) + '.jpg'
  dst = os.path.join(data_dir1, dst)
  os.rename(src,dst)
  i += 1

i = 0
for name in file_names2:
  src = os.path.join(data_dir2, name)
  dst = '분노' + str(i) + '.jpg'
  dst = os.path.join(data_dir2, dst)
  os.rename(src,dst)
  i += 1

i = 0
for name in file_names3:
  src = os.path.join(data_dir3, name)
  dst = '불안' + str(i) + '.jpg'
  dst = os.path.join(data_dir3, dst)
  os.rename(src,dst)
  i += 1

i = 0
for name in file_names4:
  src = os.path.join(data_dir4, name)
  dst = '상처' + str(i) + '.jpg'
  dst = os.path.join(data_dir4, dst)
  os.rename(src,dst)
  i += 1

i = 0
for name in file_names5:
  src = os.path.join(data_dir5, name)
  dst = '슬픔' + str(i) + '.jpg'
  dst = os.path.join(data_dir5, dst)
  os.rename(src,dst)
  i += 1

i = 0
for name in file_names6:
  src = os.path.join(data_dir6, name)
  dst = '중립' + str(i) + '.jpg'
  dst = os.path.join(data_dir6, dst)
  os.rename(src,dst)
  i += 1

listdir = os.listdir(data_dir)
listdir1 = os.listdir(data_dir1)
listdir2 = os.listdir(data_dir2)
listdir3 = os.listdir(data_dir3)
listdir4 = os.listdir(data_dir4)
listdir5 = os.listdir(data_dir5)
listdir6 = os.listdir(data_dir6)

print('total training 기쁨 images:', len(listdir))
print('total training 당황 images:', len(listdir1))
print('total training 분노 images:', len(listdir2))
print('total training 불안 images:', len(listdir3))
print('total training 상처 images:', len(listdir4))
print('total training 슬픔 images:', len(listdir5))
print('total training 중립 images:', len(listdir6))
```
